```python
# 5人のキャラクターの攻撃力データ
attack_powers = [120, 450, 310, 580, 240]

# 合計値を計算
total_power = sum(attack_powers)

# 平均値を計算
average_power = total_power / len(attack_powers)

print(f"パーティーの合計攻撃力: {total_power}")
print(f"パーティーの平均攻撃力: {average_power}")
```
