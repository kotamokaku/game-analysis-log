```python
import pandas as pd

# 1. 架空のデータを作る（ユーザーID、プレイ時間、課金有無）
data = {
    'user_id': [1, 2, 3, 4, 5],
    'play_time_min': [120, 15, 250, 45, 100],
    'is_charged': [True, False, True, False, False]
}

# 2. Pandasの「データフレーム」という形に変換する
df = pd.DataFrame(data)

# 3. 中身を確認する (head)
print("--- データの先頭を確認 ---")
print(df.head())

# 4. 統計的な要約を確認する (describe)
print("\n--- データの統計的な特徴を確認 ---")
print(df.describe())

# 5. 特定の条件（課金ユーザーだけ）で絞り込む
charged_users = df[df['is_charged'] == True]
print("\n--- 課金ユーザーのみの抽出 ---")
print(charged_users)
```

#### このデータについての第一印象
- **構成**: 行数は5件、列は3個（user_id, play_time_min, is_charged）。
- **属性**: 数値データは play_time_min、カテゴリデータは is_charged。
- **違和感**: プレイ時間の平均は106分だが、最大値（250分）が平均の2倍以上。一部のヘビーユーザーが平均を押し上げている。
- **分析できそうなこと**: 課金の有無によって、プレイ時間にどれくらい明確な差が出るのかを比較したい。

Day2：仮説
「課金をしているプレイヤーほど、1日あたりのプレイ時間が長い傾向にある」
 （理由：昨日の数値で、最大値250分という突出した数字があり、これが課金ユーザーである可能性が高いと考えたため）

#### 分析の前提条件と限界（セルフレビュー）
- **サンプルサイズの制約**: サンプル数が5件と極めて少ないため、この結果を全ユーザーに一般化することはできない。
- **特定データの影響**: 「250分」という1つの突出したデータが平均値を大きく引き上げており、平均値だけでは実態を見誤るリスクがある。
- **視点の切り替え**: 平均値ではなく「中央値（100分）」を基準に据えると、ユーザーのプレイ時間の印象は大きく変わる可能性がある。
